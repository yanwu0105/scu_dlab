```python
month_list = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
ptt_datatime = "Tue Oct 18 23:22:05 2016"

ptt_split_list = ptt_datatime.split(" ")
print(ptt_split_list)

ptt_month = ptt_split_list[1]
print(ptt_month)

for index in range(0,len(month_list)):
    ptt_month = ptt_month.replace(month_list[index],str(index+1))
print (ptt_month)


ptt_split_list.remove('Oct')
ptt_split_list.insert(1,ptt_month)
print(ptt_split_list)
print(" ".join(ptt_split_list))
```

```python
s = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."
x = []

for i in s:
    if ord(i) in range(ord('a'),(ord('z')+1)):
        if i == 'y':
            x.append(chr(ord(i)-24))
        elif i == 'z':
            x.append(chr(ord(i)-24))
        else:
            x.append(chr(ord(i)+2))
    else:
        x.append(i)
print(''.join(x))    
```
