```python
#判斷是否為回文
s = "上海自來水來自海上"
s_front = ""
s_back = ""
half_len_of_s = int(len(s)/2)
for i in range(0,half_len_of_s):
    s_front+=s[i]
for i in range(len(s)-1,half_len_of_s,-1):
    s_back+=s[i]
    
if s_front==s_back:
    print(("「%s」是回文字串")%(s))
else:
    print(("「%s」不是回文字串")%(s))
```

```python
want = []
dontwant = []
allnum = (range(100,1000))
for i in allnum:
    aaa = str(i)
    if aaa[0]==aaa[1]==aaa[2]:
        dontwant.append(i)
    elif aaa[0]==aaa[1] or aaa[0]==aaa[2] or aaa[1]==aaa[2]:
        dontwant.append(i)
    else:
        want.append(i)
print(want) 
print(len(want))
    
```

```python
want = []
dontwant = []
allnum = (range(100,1000))
for i in allnum:
    if int((i/100)) == int(((i%100)/10)) == int((i%10)):
        dontwant.append(i)
    elif int((i/100)) == int(((i%100)/10)):
        dontwant.append(i)
    elif int((i/100)) == int((i%10)):
        dontwant.append(i)
    elif int(((i%100)/10)) == int((i%10)):
        dontwant.append(i)
    else:
        want.append(i)
    
print(want) 
print(len(want))
```
